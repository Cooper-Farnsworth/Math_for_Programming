```python
import cmath
import math
import numpy as np
import mpmath as mp
from cmath import phase
from math import pi
import matplotlib.pyplot as plt

import sympy as sp
sp.init_printing()

def polar_form(z):
    r = abs(z)
    arg = cmath.phase(z)
    return r, arg

def trig_str(z, digits=6):
    r, arg = polar_form(z)
    return f"{r:.{digits}f} * (cos({arg:.{digits}f}) + i*sin({arg:.{digits}f}))"

def exp_str(z, digits=6):
    r, arg = polar_form(z)
    return f"{r:.{digits}f} * exp(i*{arg:.{digits}f})"

def alg_str(z, digits=6):
    return f"{z.real:.{digits}f} {'+' if z.imag>=0 else '-'} {abs(z.imag):.{digits}f}i"
```

#2.1

```python
import cmath, math
sqrt3 = math.sqrt(3)

# a)
za = (5 + 1j) * (7 - 6j) / (3 + 1j)
# b)
zb = (1 + 1j)**5 / (1 - 1j)**3
# c)
zc = ((-1 + 1j*sqrt3)**15) / ((1 - 1j)**20) + ((-1 - 1j*sqrt3)**15) / ((1 + 1j)**20)

def print_all_forms(label, z):
    print(f"{label}:")
    print("  Алгебраическая:", alg_str(z))
    print("  Тригонометрическая:", trig_str(z))
    print("  Показательная:", exp_str(z))
    print()

print_all_forms("a", za)
print_all_forms("b", zb)
print_all_forms("c", zc)
```

```python
w = 1 + 1j
R = abs(w)
argw = cmath.phase(w)
roots = []
for k in range(8):
    root = R**(1/8) * cmath.exp(1j*(argw + 2*math.pi*k)/8)
    roots.append(root)

for k, z in enumerate(roots):
    print(f"z[{k}] = {alg_str(z)} ; r={abs(z):.6f} ; arg={cmath.phase(z):.6f}")

xs = [z.real for z in roots]
ys = [z.imag for z in roots]
plt.figure(figsize=(5,5))
plt.axhline(0, lw=0.5)
plt.axvline(0, lw=0.5)
plt.scatter(xs, ys, s=50)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Корни уравнения z^8 = 1 + i")
plt.grid(True, ls=":")
plt.show()
```

#2.3

```python
def n_roots_of(z, n):
    R = abs(z)
    arg = cmath.phase(z)
    roots = []
    for k in range(n):
        roots.append(R**(1/n) * cmath.exp(1j*(arg + 2*math.pi*k)/n))
    return roots

# a)
za = 1+0j
ra = n_roots_of(za, 8)

# b)
zb = -18*(1 + 1j*math.sqrt(3))
rb = n_roots_of(zb, 4)

# c)
zc = (1 - 5j)/(1 + 1j) - 5*(1 + 2j)/(2 - 1j) + 11
rc = n_roots_of(zc, 3)

# d)
zd = (1 - 1j)
rd = n_roots_of(zd, 9)

def print_roots(label, roots):
    print(f"{label}: число корней = {len(roots)}")
    for k, z in enumerate(roots):
        print(f"  {k}: {alg_str(z)} ; r={abs(z):.6f} ; arg={cmath.phase(z):.6f}")
    print()

print_roots("a) 8-я корень из 1", ra)
print_roots("b) 4-я корень из -18*(1+i*sqrt(3))", rb)
print_roots("c) 3-я корень из (1-5i)/(1+i) - 5*(1+2i)/(2-i) + 11", rc)
print_roots("d) 9-я корень из (1 - i)", rd)

def plot_roots(roots, title):
    xs = [z.real for z in roots]
    ys = [z.imag for z in roots]
    plt.figure(figsize=(5,5))
    plt.axhline(0, lw=0.5)
    plt.axvline(0, lw=0.5)
    plt.scatter(xs, ys, s=30)
    plt.title(title)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.grid(True, ls=":")
    plt.show()

plot_roots(ra, "a) 8-я корень из 1")
plot_roots(rb, "b) 4-я корень из -18*(1+i√3)")
plot_roots(rc, "c) 3-я корень из сложного выражения")
plot_roots(rd, "d) 9-я корень из (1 - i)")
```

```python
X = np.linspace(-3.5, 3.5, 700)
Y = np.linspace(-3.5, 3.5, 700)
XX, YY = np.meshgrid(X, Y)
Z = XX + 1j*YY

def show_region(mask, title):
    plt.figure(figsize=(5,5))
    plt.imshow(mask[::-1, :], extent=[X.min(), X.max(), Y.min(), Y.max()], alpha=0.9)
    plt.axhline(0, lw=0.5)
    plt.axvline(0, lw=0.5)
    plt.title(title)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

mask_a = (np.abs(Z - 1j) + np.abs(Z + 1j)) < 4
show_region(mask_a, "a) |z − i| + |z + i| < 4")

mask_b = (XX + YY) < math.sqrt(2)
show_region(mask_b, "b) Re(z(1−i)) < √2  (т.е. x + y < √2)")

eps = 1e-2
mask_c = (np.abs(np.abs(Z + 1) - np.abs(Z + 1j)) < eps) & (np.abs(np.abs(Z + 1) - np.abs(Z - 1j)) < eps)
show_region(mask_c, "c) |z + 1| = |z + i| = |z − i| (прибл.)")

def arg_complex(W):
    return np.angle(W)

W_d = 1j * (-Z) / (Z + 1j)
mask_d = (~np.isclose(Z + 1j, 0)) & (arg_complex(W_d) > 0) & (arg_complex(W_d) < (np.pi/2))
show_region(mask_d, "d) 0 < arg(i −z /(z+i)) < π/2")

W_e = ((2 - 1j) * Z) / (Z + 3j)
arg_e = arg_complex(W_e)
mask_e = (arg_e >= -2*np.pi/3) & (arg_e <= -5*np.pi/6)
show_region(mask_e, "e) −2π/3 ≤ arg( (2−i) z /(z+3i) ) ≤ −5π/6")

W_f = (3 - Z) / (Z + 1j)
arg_f = arg_complex(W_f)
mask_f = (arg_f >= -2*np.pi/3) & (arg_f <= -np.pi/3)
show_region(mask_f, "f) −2π/3 ≤ arg( (3−z)/(z+i) ) ≤ −π/3")
```

```python
def circle_from_general(A, B, C):
    assert A > 0, "Требуется A>0"
    z0 = -B / A
    R2 = (abs(B)**2)/A**2 - C/A
    return z0, math.sqrt(R2)

A = 2.0
B = 1 + 2j
C = -1.5
z0, R = circle_from_general(A,B,C)
print("Центр:", alg_str(z0), "Радиус:", f"{R:.6f}")
```

```python
mp.mp.dps = 30

def C_w_pow_z(z, w):
    return mp.gamma(z+1) / (mp.gamma(w+1) * mp.gamma(z-w+1))

z = 1 + 1j
w = 1 - 1j
val = C_w_pow_z(z, w)
print("a) C_{1−i}^{1+i} =", val)
print("b) Arg(C_{1−i}^{1+i}) =", mp.arg(val))
angles = [ -3*math.pi/4, -math.pi/2, -math.pi/4, 0, math.pi/4, math.pi/2, 3*math.pi/4 ]
r = np.linspace(0, 3, 300)

plt.figure(figsize=(6,6))
for a in angles:
    x = r*np.cos(a)
    y = r*np.sin(a)
    plt.plot(x, y, lw=1)
plt.axhline(0, lw=0.5, ls=":")
plt.axvline(0, lw=0.5, ls=":")
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Линии уровня Arg(C_z^1) = Arg(z) = const")
plt.grid(True, ls=":")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def julia_quadratic(c, N=800, itmax=300, bound=4, xlim=(-2,2), ylim=(-2,2)):
    xs = np.linspace(xlim[0], xlim[1], N)
    ys = np.linspace(ylim[0], ylim[1], N)
    img = np.zeros((N,N))
    for ix, x in enumerate(xs):
        for iy, y in enumerate(ys):
            z = x + 1j*y
            it = 0
            while abs(z) <= bound and it < itmax:
                z = z*z + c
                it += 1
            img[iy, ix] = it
    return img

def julia_rational(f, N=800, itmax=300, bound=20, xlim=(-2,2), ylim=(-2,2)):
    xs = np.linspace(xlim[0], xlim[1], N)
    ys = np.linspace(ylim[0], ylim[1], N)
    img = np.zeros((N,N))
    for ix, x in enumerate(xs):
        for iy, y in enumerate(ys):
            z = x + 1j*y
            it = 0
            try:
                while abs(z) <= bound and it < itmax:
                    z = f(z)
                    it += 1
            except ZeroDivisionError:
                it = itmax
            img[iy, ix] = it
    return img

for c in [ -0.1 + 0.65j, 0.28 + 0.009j ]:
    img = julia_quadratic(c, N=500, itmax=400, bound=10, xlim=(-2.1,2.1), ylim=(-2.1,2.1))
    plt.figure(figsize=(6,6))
    plt.imshow(img, origin='lower', extent=[-2.1,2.1,-2.1,2.1])
    plt.title(f"Julia f_c(z)=z^2+c, c={c}")
    plt.axis('off')
    plt.show()
def f1(z):
    return (1j*z**5 - 1) / (z**5 - 1j)

img1 = julia_rational(f1, N=500, itmax=200, bound=30, xlim=(-2,2), ylim=(-2,2))
plt.figure(figsize=(6,6))
plt.imshow(img1, origin='lower', extent=[-2,2,-2,2])
plt.title("Julia for f(z) = (i z^5 − 1)/(z^5 − i)")
plt.axis('off')
plt.show()

def f2(z):
    return (z**2 + 1j) / (1 - z**2)

img2 = julia_rational(f2, N=500, itmax=200, bound=30, xlim=(-2,2), ylim=(-2,2))
plt.figure(figsize=(6,6))
plt.imshow(img2, origin='lower', extent=[-2,2,-2,2])
plt.title("Julia for f(z) = (z^2 + i)/(1 − z^2)")
plt.axis('off')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

def stereographic_to_sphere(z):
    x, y = z.real, z.imag
    denom = 1 + x*x + y*y
    X = 2*x/denom
    Y = 2*y/denom
    Z = (x*x + y*y - 1)/denom
    return X, Y, Z

def plot_curve_on_sphere(pts, title=""):
    XYZ = np.array([stereographic_to_sphere(p) for p in pts])
    fig = plt.figure(figsize=(6,6))
    ax = fig.add_subplot(111, projection='3d')
    u = np.linspace(0, 2*np.pi, 60)
    v = np.linspace(0, np.pi, 60)
    xs = np.outer(np.cos(u), np.sin(v))
    ys = np.outer(np.sin(u), np.sin(v))
    zs = np.outer(np.ones_like(u), np.cos(v))
    ax.plot_surface(xs, ys, zs, alpha=0.15, linewidth=0)
    ax.plot(XYZ[:,0], XYZ[:,1], XYZ[:,2], lw=2)
    ax.set_title(title)
    ax.set_box_aspect([1,1,1])
    plt.show()

t = np.linspace(-5, 5, 2000)
line = t + 1j*(0.5*t + 0.0)
plot_curve_on_sphere(line, "Стереографическое изображение прямой")

phi = np.linspace(0, 2*np.pi, 1000)
circle = 1.0*np.cos(phi) + 1j*1.0*np.sin(phi)
plot_curve_on_sphere(circle, "Стереографическое изображение окружности")

parab = t + 1j*(t**2)
plot_curve_on_sphere(parab, "Стереографическое изображение параболы")

tt = np.linspace(-5, -0.2, 1000)
tt2 = np.linspace(0.2, 5, 1000)
hyp = np.concatenate([tt + 1j*(1/tt), tt2 + 1j*(1/tt2)])
plot_curve_on_sphere(hyp, "Стереографическое изображение гиперболы")

a, b = 0.25, 0.20
phi = np.linspace(-6*np.pi, 6*np.pi, 5000)
r = a * np.exp(b*phi)
spiral = r*np.cos(phi) + 1j*r*np.sin(phi)
plot_curve_on_sphere(spiral, "Стереографическое изображение логарифмической спирали")
```
